#  Задача 1

```python
df['date'] = df['date'].str.split('T', expand=True)[0]
df
```

```python
df['date'] = pd.to_datetime(df['date'], format = 'ISO8601')
df
```

```python
df.info()
```

```python
df['date'] = df['date'].astype('str')
df
```

```python
df.dropna(inplace=True)
new = df['date'].str.split('-', n=2, expand=True)
df['date_year']=new[0]
df['date_month']=new[1]
df['date_day']=new[2]
#df.drop(columns=['Memory'], inplace=True)
df
```

```python
df.info()
```

```python
plt.hist(df['price'], color = 'green', edgecolor = 'black',
         bins = int(180/5))

sns.distplot(df['price'],  bins=int(180/5), hist_kws={'edgecolor':'black'})
plt.title('1.2. Изучите стоимости недвижимости')
plt.xlabel('')
plt.ylabel('');
```

```python
plt.figure(figsize=(6, 4))

sns.histplot(df['sqft_lot'], bins=15)

plt.title('1.3. Изучите распределение квадратуры жилой')
plt.xlabel('')
plt.ylabel('');
```

```python
plt.figure(figsize=(6, 4))

sns.histplot(df['date_year'], bins=30)

plt.title('1.4. Изучите распределение года постройки')
plt.xlabel('')
plt.ylabel('');
```

```python
df['year_and_month']=df['date_year'] + df['date_month']
df.sample(20)
```

```python
plt.figure(figsize=(7, 4))

sns.histplot(df['year_and_month'], bins=24, color = 'pink', edgecolor = 'black')

plt.title('1.4. Изучите распределение года постройки')
plt.xlabel('Date')
plt.ylabel('')
plt.xticks(rotation=30);
```

```python
counts = df['view'].value_counts()
    
plt.figure(figsize=(8, 6))    
plt.title('view')
sns.barplot(x=counts.index, y=counts.values)
    
plt.show()
```

Вывод: Следуя из данного графика, можно сказать, что квартир с видом №0  значительно  больше чем квартир с другим видом. 

```python
df['bedrooms'] = df['bedrooms'].astype('float64')
df['floors'] = df['floors'].astype('int64')
df
```

```python
view_price = df.pivot_table(index = 'view', values = 'price', aggfunc = ['mean', 'count', 'median'])
view_price.columns = ['mean', 'count', 'median']
view_price.query('count > 50').plot(y = 'median')

#view_price.query('count > 50').sort_values('median', ascending = False)

df['view'].corr(df['price']);
```

```python
pivot_table_rooms = df.pivot_table(index = 'bedrooms', values = 'price', aggfunc = ['mean', 'count', 'median'])
pivot_table_rooms.columns = ['mean', 'count', 'median']
pivot_table_rooms.query('count > 50').plot(y = 'median')

pivot_table_rooms.query('count > 50').sort_values('median', ascending = False)

df['bedrooms'].corr(df['price']);
```

Вывод: чем больше комнат тем дороже стоимость квартиры.
